**Download company financials - fundamental and historical price data**
*   [Yfinance](https://pypi.org/project/yfinance/)


```
symbol = 'TRIP'
ticker = yf.Ticker(symbol)
```

**Download Fundamental Data**

```
# get stock info
ticker.info
```

```
print('marketCap:', ticker.info['marketCap'], 'sector:', ticker.info['sector'])
```

**Download Market Data**

```
data = yf.download(symbol, interval = '1m', start="2020-05-20", end="2020-05-21")
data.head(20)
```
